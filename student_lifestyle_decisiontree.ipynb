```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn import tree
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('student_lifestyle_dataset.csv')
```

```python
df.head()
```

```python
features = ['Study_Hours_Per_Day', 'Extracurricular_Hours_Per_Day', 'Sleep_Hours_Per_Day', 'Social_Hours_Per_Day', 'Physical_Activity_Hours_Per_Day', 'Stress_Level', 'Gender']
```

```python
df = df.replace({'Low' : 0, 'Moderate' : 1, 'High' : 2, 'Male' : 0, 'Female' : 1})
```

```python
x = df[features]
y = df['Grades']
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.3, random_state = 0)
model = DecisionTreeRegressor(random_state = 0)
```

```python
model.fit(train_x, train_y)
```

```python
predictions = model.predict(test_x)
```

```python
model.score(test_x, test_y)
```

```python
mean_squared_error(test_y, predictions)
```

```python
mean_absolute_error(test_y, predictions)
```

```python
model.get_depth()
```

```python
gs = GridSearchCV(DecisionTreeRegressor(random_state=0), param_grid = {'max_depth' : range(2, 21)}, scoring = 'r2')
```

```python
gs.fit(train_x, train_y)
```

```python
model2 = DecisionTreeRegressor(random_state = 0, max_depth = 3)
```

```python
model2.fit(train_x, train_y)
```

```python
predictions2 = model2.predict(test_x)
```

```python
model2.score(test_x, test_y)
```

```python
mean_squared_error(test_y, predictions2)
```

```python
mean_absolute_error(test_y, predictions2)
```
