```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv('student_lifestyle_dataset.csv')
```

```python
df
```

```python
df.info()
```

```python
features = ['Study_Hours_Per_Day', 'Extracurricular_Hours_Per_Day', 'Sleep_Hours_Per_Day', 'Social_Hours_Per_Day', 'Physical_Activity_Hours_Per_Day', 'Stress_Level', 'Gender']
```

```python
df[features]
```

```python
df = df.replace({'Low' : 0, 'Moderate' : 1, 'High' : 2, 'Male' : 0, 'Female' : 1})
```

```python
df
```

```python
x = df[features]
y = df['Grades']
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.3, random_state = 0)
model = LinearRegression()
```

```python
model.fit(train_x, train_y)
```

```python
predictions = model.predict(test_x)
```

```python
model.score(test_x, test_y)
```

```python
mean_squared_error(test_y, predictions)
```

```python
mean_absolute_error(test_y, predictions)
```
